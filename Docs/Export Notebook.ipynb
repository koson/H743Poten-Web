````python
import subprocess
import os
import webbrowser
from pathlib import Path

# Get current notebook path
notebook_path = Path(__file__).parent if '__file__' in globals() else Path.cwd()
current_notebook = notebook_path / "Export Notebook.ipynb"

print(f"Current notebook: {current_notebook}")
print(f"Notebook exists: {current_notebook.exists()}")

# Export to HTML (‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡πÉ‡∏ä‡πâ Pandoc)
if current_notebook.exists():
    try:
        # Export ‡πÄ‡∏õ‡πá‡∏ô HTML
        html_output = current_notebook.with_suffix('.html')
        
        # Run nbconvert to HTML
        result = subprocess.run([
            'python', '-m', 'jupyter', 'nbconvert', 
            '--to', 'html',
            '--output', str(html_output),
            str(current_notebook)
        ], capture_output=True, text=True, cwd=notebook_path)
        
        if result.returncode == 0:
            print(f"‚úÖ Successfully exported to: {html_output}")
            print(f"� HTML file size: {html_output.stat().st_size} bytes")
            
            # ‡πÄ‡∏õ‡∏¥‡∏î‡πÉ‡∏ô‡πÄ‡∏ö‡∏£‡∏≤‡∏ß‡πå‡πÄ‡∏ã‡∏≠‡∏£‡πå
            webbrowser.open(f"file://{html_output.absolute()}")
            print("� Opened in browser - use Ctrl+P to print as PDF")
            
        else:
            print(f"‚ùå Export failed:")
            print(f"STDOUT: {result.stdout}")
            print(f"STDERR: {result.stderr}")
            
    except Exception as e:
        print(f"‚ùå Error during export: {e}")
else:
    print("‚ùå Notebook file not found")

# ‡∏ß‡∏¥‡∏ò‡∏µ‡∏ó‡∏≤‡∏á‡πÄ‡∏•‡∏∑‡∏≠‡∏Å: ‡πÉ‡∏ä‡πâ nbconvert ‡πÇ‡∏î‡∏¢‡∏ï‡∏£‡∏á‡πÉ‡∏ô command line
print("\n" + "="*50)
print("üìã Manual export commands:")
print("="*50)
print("1. Export to HTML:")
print('   jupyter nbconvert --to html "Export Notebook.ipynb"')
print("\n2. Export to PDF (‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ Pandoc + LaTeX):")
print('   jupyter nbconvert --to pdf "Export Notebook.ipynb"')
print("\n3. Export to slides:")
print('   jupyter nbconvert --to slides "Export Notebook.ipynb"')
````

### **5. üê≥ Batch Export Script**

‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÑ‡∏ü‡∏•‡πå `export_all.py`:
```python
#!/usr/bin/env python3
import subprocess
import sys
from pathlib import Path

def export_notebook(notebook_path, formats=['html', 'markdown', 'python']):
    """Export notebook ‡∏´‡∏•‡∏≤‡∏¢‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö"""
    notebook = Path(notebook_path)
    
    for fmt in formats:
        cmd = ['jupyter', 'nbconvert', '--to', fmt, str(notebook)]
        try:
            subprocess.run(cmd, check=True)
            print(f"‚úÖ Exported to {fmt}")
        except subprocess.CalledProcessError:
            print(f"‚ùå Failed to export to {fmt}")

if __name__ == "__main__":
    if len(sys.argv) > 1:
        export_notebook(sys.argv[1])
    else:
        print("Usage: python export_all.py notebook.ipynb")
```

````python
# üîß **WORKING EXPORT SOLUTIONS**

import json
import os
from pathlib import Path
import subprocess
import sys

class NotebookExportFixer:
    """‡πÅ‡∏Å‡πâ‡∏õ‡∏±‡∏ç‡∏´‡∏≤ export ‡πÅ‡∏•‡∏∞‡πÉ‡∏´‡πâ alternative methods"""
    
    def __init__(self, notebook_path):
        self.notebook_path = Path(notebook_path)
        self.notebook_name = self.notebook_path.stem
        
    def check_dependencies(self):
        """‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö dependencies ‡∏ó‡∏µ‡πà‡∏à‡∏≥‡πÄ‡∏õ‡πá‡∏ô"""
        print("üîç Checking Export Dependencies...")
        print("=" * 40)
        
        dependencies = {
            "jupyter": {"cmd": "jupyter --version", "install": "pip install jupyter"},
            "nbconvert": {"cmd": "jupyter nbconvert --version", "install": "pip install nbconvert"},
            "pandoc": {"cmd": "pandoc --version", "install": "Download from https://pandoc.org/"},
            "python": {"cmd": "python --version", "install": "Already available"}
        }
        
        available = []
        missing = []
        
        for name, info in dependencies.items():
            try:
                result = subprocess.run(info["cmd"].split(), 
                                     capture_output=True, text=True, check=True)
                print(f"  ‚úÖ {name}: Available")
                available.append(name)
            except (subprocess.CalledProcessError, FileNotFoundError):
                print(f"  ‚ùå {name}: Missing")
                print(f"     Install: {info['install']}")
                missing.append(name)
        
        return available, missing
    
    def export_to_html_manual(self):
        """Manual HTML export (‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡πÉ‡∏ä‡πâ nbconvert)"""
        print("\nüåê Manual HTML Export...")
        
        try:
            # ‡∏≠‡πà‡∏≤‡∏ô notebook
            with open(self.notebook_path, 'r', encoding='utf-8') as f:
                notebook_data = json.load(f)
            
            # ‡∏™‡∏£‡πâ‡∏≤‡∏á HTML basic
            html_content = f"""
<!DOCTYPE html>
<html>
<head>
    <title>{self.notebook_name}</title>
    <meta charset="utf-8">
    <style>
        body {{ font-family: Arial, sans-serif; margin: 40px; }}
        .cell {{ margin: 20px 0; border-left: 3px solid #ddd; padding-left: 20px; }}
        .markdown {{ }}
        .code {{ background-color: #f8f8f8; padding: 10px; border-radius: 4px; }}
        .output {{ background-color: #fff; border: 1px solid #ddd; padding: 10px; }}
        pre {{ white-space: pre-wrap; }}
        h1 {{ color: #333; }}
        h2 {{ color: #666; }}
        h3 {{ color: #999; }}
    </style>
</head>
<body>
    <h1>{self.notebook_name}</h1>
    <p><em>Exported on {os.path.basename(__file__)}</em></p>
"""
            
            # Process cells
            for i, cell in enumerate(notebook_data.get('cells', [])):
                cell_type = cell.get('cell_type', 'code')
                source = ''.join(cell.get('source', []))
                
                html_content += f'\\n    <div class="cell {cell_type}">'
                
                if cell_type == 'markdown':
                    # Convert markdown to HTML (basic)
                    html_content += f'\\n        <div class="markdown">{source}</div>'
                elif cell_type == 'code':
                    html_content += f'\\n        <div class="code"><pre>{source}</pre></div>'
                    
                    # Add outputs if available
                    outputs = cell.get('outputs', [])
                    if outputs:
                        html_content += '\\n        <div class="output">'
                        for output in outputs:
                            if 'text' in output:
                                text_output = ''.join(output['text'])
                                html_content += f'<pre>{text_output}</pre>'
                        html_content += '</div>'
                
                html_content += '\\n    </div>'
            
            html_content += """
</body>
</html>
"""
            
            # Save HTML file
            output_file = self.notebook_path.parent / f"{self.notebook_name}_manual.html"
            with open(output_file, 'w', encoding='utf-8') as f:
                f.write(html_content)
            
            print(f"  ‚úÖ Manual HTML created: {output_file}")
            return str(output_file)
            
        except Exception as e:
            print(f"  ‚ùå Manual HTML failed: {e}")
            return None
    
    def export_to_markdown_manual(self):
        """Manual Markdown export"""
        print("\nüìù Manual Markdown Export...")
        
        try:
            # ‡∏≠‡πà‡∏≤‡∏ô notebook
            with open(self.notebook_path, 'r', encoding='utf-8') as f:
                notebook_data = json.load(f)
            
            # ‡∏™‡∏£‡πâ‡∏≤‡∏á Markdown content
            md_content = f"# {self.notebook_name}\\n\\n"
            md_content += f"*Exported manually from Jupyter Notebook*\\n\\n"
            
            # Process cells
            for i, cell in enumerate(notebook_data.get('cells', [])):
                cell_type = cell.get('cell_type', 'code')
                source = ''.join(cell.get('source', []))
                
                if cell_type == 'markdown':
                    md_content += f"{source}\\n\\n"
                elif cell_type == 'code':
                    md_content += f"```python\\n{source}\\n```\\n\\n"
                    
                    # Add outputs
                    outputs = cell.get('outputs', [])
                    if outputs:
                        md_content += "**Output:**\\n```\\n"
                        for output in outputs:
                            if 'text' in output:
                                text_output = ''.join(output['text'])
                                md_content += f"{text_output}\\n"
                        md_content += "```\\n\\n"
            
            # Save Markdown file
            output_file = self.notebook_path.parent / f"{self.notebook_name}_manual.md"
            with open(output_file, 'w', encoding='utf-8') as f:
                f.write(md_content)
            
            print(f"  ‚úÖ Manual Markdown created: {output_file}")
            return str(output_file)
            
        except Exception as e:
            print(f"  ‚ùå Manual Markdown failed: {e}")
            return None
    
    def export_to_python_manual(self):
        """Manual Python script export"""
        print("\nüêç Manual Python Export...")
        
        try:
            # ‡∏≠‡πà‡∏≤‡∏ô notebook
            with open(self.notebook_path, 'r', encoding='utf-8') as f:
                notebook_data = json.load(f)
            
            # ‡∏™‡∏£‡πâ‡∏≤‡∏á Python content
            py_content = f"#!/usr/bin/env python3\\n"
            py_content += f'"""{self.notebook_name}\\n'
            py_content += f"Exported from Jupyter Notebook\\n"
            py_content += f'"""\\n\\n'
            
            # Process cells
            for i, cell in enumerate(notebook_data.get('cells', [])):
                cell_type = cell.get('cell_type', 'code')
                source = ''.join(cell.get('source', []))
                
                if cell_type == 'markdown':
                    # Convert markdown to comments
                    lines = source.split('\\n')
                    py_content += "# " + "\\n# ".join(lines) + "\\n\\n"
                elif cell_type == 'code':
                    py_content += f"# Cell {i+1}\\n"
                    py_content += f"{source}\\n\\n"
            
            # Save Python file
            output_file = self.notebook_path.parent / f"{self.notebook_name}_manual.py"
            with open(output_file, 'w', encoding='utf-8') as f:
                f.write(py_content)
            
            print(f"  ‚úÖ Manual Python script created: {output_file}")
            return str(output_file)
            
        except Exception as e:
            print(f"  ‚ùå Manual Python export failed: {e}")
            return None
    
    def try_nbconvert_with_fallbacks(self, format_type='html'):
        """‡∏•‡∏≠‡∏á nbconvert ‡∏´‡∏•‡∏≤‡∏¢‡∏ß‡∏¥‡∏ò‡∏µ"""
        print(f"\\n‚ö° Trying nbconvert for {format_type}...")
        
        commands = [
            f"jupyter nbconvert --to {format_type} '{self.notebook_path}'",
            f"python -m nbconvert --to {format_type} '{self.notebook_path}'",
            f"nbconvert --to {format_type} '{self.notebook_path}'"
        ]
        
        for cmd in commands:
            try:
                print(f"  üîÑ Trying: {cmd}")
                result = subprocess.run(cmd, shell=True, capture_output=True, text=True, check=True)
                print(f"  ‚úÖ Success with: {cmd}")
                return True
            except subprocess.CalledProcessError as e:
                print(f"  ‚ùå Failed: {e}")
                continue
        
        print(f"  ‚ùå All nbconvert methods failed for {format_type}")
        return False
    
    def export_all_methods(self):
        """‡∏•‡∏≠‡∏á‡∏ó‡∏∏‡∏Å‡∏ß‡∏¥‡∏ò‡∏µ‡∏Å‡∏≤‡∏£ export"""
        print(f"üöÄ COMPREHENSIVE EXPORT ATTEMPT")
        print(f"Notebook: {self.notebook_name}")
        print("=" * 50)
        
        # 1. Check dependencies
        available, missing = self.check_dependencies()
        
        # 2. Try nbconvert if available
        results = {}
        if 'nbconvert' in available:
            print("\\n‚ö° Trying nbconvert methods...")
            results['html_nbconvert'] = self.try_nbconvert_with_fallbacks('html')
            results['markdown_nbconvert'] = self.try_nbconvert_with_fallbacks('markdown')
            results['python_nbconvert'] = self.try_nbconvert_with_fallbacks('python')
        
        # 3. Manual methods (always work)
        print("\\nüîß Trying manual methods...")
        results['html_manual'] = self.export_to_html_manual() is not None
        results['markdown_manual'] = self.export_to_markdown_manual() is not None
        results['python_manual'] = self.export_to_python_manual() is not None
        
        # 4. Summary
        print(f"\\nüìä EXPORT SUMMARY:")
        print("=" * 30)
        successful = 0
        for method, success in results.items():
            status = "‚úÖ Success" if success else "‚ùå Failed"
            print(f"  {method}: {status}")
            if success:
                successful += 1
        
        print(f"\\nüéØ Success Rate: {successful}/{len(results)} methods")
        
        if successful > 0:
            print("\\nüí° Recommendations:")
            if results.get('html_manual', False):
                print("  üìÑ Use manual HTML for sharing")
            if results.get('markdown_manual', False):
                print("  üìù Use manual Markdown for documentation")
            if results.get('python_manual', False):
                print("  üêç Use manual Python for code extraction")
        else:
            print("\\n‚ùå All export methods failed!")
            print("üí° Try copying content manually or check file permissions")
        
        return results

# üéØ ‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô‡∏Å‡∏±‡∏ö notebook ‡∏õ‡∏±‡∏à‡∏à‡∏∏‡∏ö‡∏±‡∏ô
current_notebook = r"d:\\GitHubRepos\\__Potentiostat\\poten-2025\\H743Poten\\H743Poten-Web\\Docs\\Algorithm_Analysis_DeepCV.ipynb"

print("üîß NOTEBOOK EXPORT TROUBLESHOOTER")
print("=" * 50)

# ‡∏™‡∏£‡πâ‡∏≤‡∏á fixer ‡πÅ‡∏•‡∏∞‡∏•‡∏≠‡∏á‡∏ó‡∏∏‡∏Å‡∏ß‡∏¥‡∏ò‡∏µ
fixer = NotebookExportFixer(current_notebook)
results = fixer.export_all_methods()

print("\\nüéâ TROUBLESHOOTING COMPLETE!")
print("üìÅ Check your notebook directory for exported files")
````

**üéØ ‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏ß‡∏¥‡∏ò‡∏µ‡∏ó‡∏µ‡πà‡πÄ‡∏´‡∏°‡∏≤‡∏∞‡∏™‡∏°‡∏Å‡∏±‡∏ö‡∏™‡∏ñ‡∏≤‡∏ô‡∏Å‡∏≤‡∏£‡∏ì‡πå‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏∏‡∏ì ‡πÅ‡∏•‡πâ‡∏ß‡∏•‡∏≠‡∏á‡∏ï‡∏≤‡∏°‡∏•‡∏≥‡∏î‡∏±‡∏ö!**

````python
# üöÄ **‡∏£‡∏±‡∏ô‡πÇ‡∏Ñ‡πâ‡∏î‡∏ô‡∏µ‡πâ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÅ‡∏Å‡πâ‡∏õ‡∏±‡∏ç‡∏´‡∏≤ Export ‡∏ó‡∏±‡∏ô‡∏ó‡∏µ!**

import os
import json
import subprocess
import sys

def fix_export_problem():
    """‡πÅ‡∏Å‡πâ‡∏õ‡∏±‡∏ç‡∏´‡∏≤ export ‡∏ó‡∏±‡∏ô‡∏ó‡∏µ!"""
    print("üîß ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡πÅ‡∏Å‡πâ‡∏õ‡∏±‡∏ç‡∏´‡∏≤ Export...")
    
    # 1. ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö dependencies
    print("\nüì¶ ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö dependencies:")
    
    try:
        import nbconvert
        print("‚úÖ nbconvert: ‡∏ï‡∏¥‡∏î‡∏ï‡∏±‡πâ‡∏á‡πÅ‡∏•‡πâ‡∏ß")
        print(f"   Version: {nbconvert.__version__}")
    except ImportError:
        print("‚ùå nbconvert: ‡πÑ‡∏°‡πà‡∏°‡∏µ - ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏ï‡∏¥‡∏î‡∏ï‡∏±‡πâ‡∏á...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", "nbconvert"])
        print("‚úÖ ‡∏ï‡∏¥‡∏î‡∏ï‡∏±‡πâ‡∏á nbconvert ‡πÄ‡∏£‡∏µ‡∏¢‡∏ö‡∏£‡πâ‡∏≠‡∏¢")
    
    # 2. ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö pandoc
    try:
        result = subprocess.run(["pandoc", "--version"], 
                              capture_output=True, text=True)
        if result.returncode == 0:
            print("‚úÖ pandoc: ‡∏ï‡∏¥‡∏î‡∏ï‡∏±‡πâ‡∏á‡πÅ‡∏•‡πâ‡∏ß")
        else:
            print("‚ùå pandoc: ‡πÑ‡∏°‡πà‡∏û‡∏ö")
            print("üí° ‡∏ï‡∏¥‡∏î‡∏ï‡∏±‡πâ‡∏á‡∏à‡∏≤‡∏Å: https://pandoc.org/installing.html")
    except FileNotFoundError:
        print("‚ùå pandoc: ‡πÑ‡∏°‡πà‡∏û‡∏ö")
        print("üí° ‡∏ï‡∏¥‡∏î‡∏ï‡∏±‡πâ‡∏á‡∏à‡∏≤‡∏Å: https://pandoc.org/installing.html")
    
    # 3. ‡∏•‡∏≠‡∏á export ‡∏á‡πà‡∏≤‡∏¢‡πÜ
    notebook_path = "Algorithm_Analysis_DeepCV.ipynb"
    if os.path.exists(notebook_path):
        print(f"\nüìì ‡∏û‡∏ö notebook: {notebook_path}")
        
        # ‡∏•‡∏≠‡∏á export HTML
        try:
            cmd = [sys.executable, "-m", "nbconvert", "--to", "html", notebook_path]
            result = subprocess.run(cmd, capture_output=True, text=True)
            
            if result.returncode == 0:
                print("‚úÖ Export HTML ‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à!")
                html_file = notebook_path.replace('.ipynb', '.html')
                print(f"üìÑ ‡πÑ‡∏ü‡∏•‡πå: {html_file}")
            else:
                print("‚ùå Export HTML ‡∏•‡πâ‡∏°‡πÄ‡∏´‡∏•‡∏ß")
                print(f"Error: {result.stderr}")
                
                # ‡∏•‡∏≠‡∏á‡∏ß‡∏¥‡∏ò‡∏µ‡∏≠‡∏∑‡πà‡∏ô
                manual_export_html(notebook_path)
                
        except Exception as e:
            print(f"‚ùå Error: {e}")
            manual_export_html(notebook_path)
    
    else:
        print(f"‚ùå ‡πÑ‡∏°‡πà‡∏û‡∏ö‡πÑ‡∏ü‡∏•‡πå: {notebook_path}")
        print("üìÅ ‡πÑ‡∏ü‡∏•‡πå‡∏õ‡∏±‡∏à‡∏à‡∏∏‡∏ö‡∏±‡∏ô:")
        for f in os.listdir('.'):
            if f.endswith('.ipynb'):
                print(f"   {f}")

def manual_export_html(notebook_path):
    """Export ‡πÄ‡∏õ‡πá‡∏ô HTML ‡πÅ‡∏ö‡∏ö manual"""
    print("\nüõ†Ô∏è ‡∏•‡∏≠‡∏á Manual Export...")
    
    try:
        with open(notebook_path, 'r', encoding='utf-8') as f:
            notebook = json.load(f)
        
        html_content = """<!DOCTYPE html>
<html>
<head>
    <title>Algorithm Analysis DeepCV</title>
    <meta charset="UTF-8">
    <style>
        body { 
            font-family: 'Segoe UI', Arial, sans-serif; 
            max-width: 1200px; 
            margin: 0 auto; 
            padding: 20px;
            line-height: 1.6;
        }
        .cell { margin: 20px 0; border-left: 3px solid #ddd; padding-left: 15px; }
        .markdown { }
        .code { 
            background: #f8f9fa; 
            padding: 15px; 
            border-radius: 5px;
            border: 1px solid #e9ecef;
        }
        pre { 
            white-space: pre-wrap; 
            margin: 0;
            font-family: 'Consolas', 'Monaco', monospace;
        }
        .output {
            background: #fff;
            border: 1px solid #dee2e6;
            padding: 10px;
            margin-top: 10px;
            border-radius: 3px;
        }
        h1, h2, h3 { color: #2c3e50; }
        .cell-type { 
            font-size: 12px; 
            color: #6c757d; 
            text-transform: uppercase;
            margin-bottom: 5px;
        }
    </style>
</head>
<body>
    <h1>üß† Algorithm Analysis DeepCV</h1>
    <p><em>Exported: """ + str(os.path.getctime(notebook_path)) + """</em></p>
"""
        
        for i, cell in enumerate(notebook['cells']):
            cell_type = cell['cell_type']
            
            html_content += f'<div class="cell">\n'
            html_content += f'<div class="cell-type">{cell_type} cell #{i+1}</div>\n'
            
            if cell_type == 'markdown':
                content = ''.join(cell['source'])
                # Simple markdown to HTML conversion
                content = content.replace('# ', '<h1>').replace('\n', '</h1>\n', 1)
                content = content.replace('## ', '<h2>').replace('\n', '</h2>\n', 1)
                content = content.replace('### ', '<h3>').replace('\n', '</h3>\n', 1)
                content = content.replace('\n\n', '</p><p>')
                content = f'<div class="markdown"><p>{content}</p></div>\n'
                html_content += content
                
            elif cell_type == 'code':
                code_content = ''.join(cell['source'])
                html_content += f'<div class="code"><pre>{code_content}</pre></div>\n'
                
                # Add outputs if any
                if 'outputs' in cell and cell['outputs']:
                    html_content += '<div class="output">'
                    for output in cell['outputs']:
                        if output.get('output_type') == 'stream':
                            text = ''.join(output.get('text', []))
                            html_content += f'<pre>{text}</pre>'
                        elif output.get('output_type') == 'execute_result':
                            if 'data' in output and 'text/plain' in output['data']:
                                text = ''.join(output['data']['text/plain'])
                                html_content += f'<pre>{text}</pre>'
                    html_content += '</div>'
            
            html_content += '</div>\n'
        
        html_content += """
</body>
</html>"""
        
        # Save HTML file
        html_file = notebook_path.replace('.ipynb', '_manual.html')
        with open(html_file, 'w', encoding='utf-8') as f:
            f.write(html_content)
        
        print(f"‚úÖ Manual Export ‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à!")
        print(f"üìÑ ‡πÑ‡∏ü‡∏•‡πå: {html_file}")
        print(f"üåê ‡πÄ‡∏õ‡∏¥‡∏î‡πÑ‡∏ü‡∏•‡πå‡πÉ‡∏ô‡πÄ‡∏ö‡∏£‡∏≤‡∏ß‡πå‡πÄ‡∏ã‡∏≠‡∏£‡πå‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏î‡∏π")
        
    except Exception as e:
        print(f"‚ùå Manual Export ‡∏•‡πâ‡∏°‡πÄ‡∏´‡∏•‡∏ß: {e}")

def simple_copy_paste_version():
    """‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÄ‡∏ß‡∏≠‡∏£‡πå‡∏ä‡∏±‡∏ô‡∏ó‡∏µ‡πà copy-paste ‡πÑ‡∏î‡πâ‡∏á‡πà‡∏≤‡∏¢"""
    notebook_path = "Algorithm_Analysis_DeepCV.ipynb"
    
    if not os.path.exists(notebook_path):
        print("‚ùå ‡πÑ‡∏°‡πà‡∏û‡∏ö‡πÑ‡∏ü‡∏•‡πå notebook")
        return
    
    try:
        with open(notebook_path, 'r', encoding='utf-8') as f:
            notebook = json.load(f)
        
        output_text = "# Algorithm Analysis DeepCV\n\n"
        
        for i, cell in enumerate(notebook['cells']):
            cell_type = cell['cell_type']
            content = ''.join(cell['source'])
            
            if cell_type == 'markdown':
                output_text += f"{content}\n\n"
            elif cell_type == 'code':
                output_text += f"```python\n{content}\n```\n\n"
        
        # Save as markdown
        md_file = notebook_path.replace('.ipynb', '_simple.md')
        with open(md_file, 'w', encoding='utf-8') as f:
            f.write(output_text)
        
        print(f"‚úÖ Markdown version ‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à!")
        print(f"üìÑ ‡πÑ‡∏ü‡∏•‡πå: {md_file}")
        print("üìã ‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ copy-paste ‡πÑ‡∏õ‡πÉ‡∏™‡πà‡πÉ‡∏ô Word/Google Docs ‡πÑ‡∏î‡πâ")
        
    except Exception as e:
        print(f"‚ùå Markdown export ‡∏•‡πâ‡∏°‡πÄ‡∏´‡∏•‡∏ß: {e}")

# üöÄ **‡∏£‡∏±‡∏ô‡πÄ‡∏•‡∏¢!**
print("üéØ **‡πÅ‡∏Å‡πâ‡∏õ‡∏±‡∏ç‡∏´‡∏≤ Export ‡∏ó‡∏±‡∏ô‡∏ó‡∏µ!**")
print("=" * 50)

fix_export_problem()
print("\n" + "=" * 50)
simple_copy_paste_version()

print("\nüéâ **‡πÄ‡∏™‡∏£‡πá‡∏à‡πÅ‡∏•‡πâ‡∏ß!**")
print("üìã **‡∏ß‡∏¥‡∏ò‡∏µ‡πÉ‡∏ä‡πâ‡πÑ‡∏ü‡∏•‡πå‡∏ó‡∏µ‡πà export ‡πÅ‡∏•‡πâ‡∏ß:**")
print("   1. HTML: ‡πÄ‡∏õ‡∏¥‡∏î‡∏î‡πâ‡∏ß‡∏¢ web browser")
print("   2. Markdown: copy-paste ‡πÑ‡∏õ Word/Google Docs")
print("   3. ‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô PDF ‡∏à‡∏≤‡∏Å browser (Print ‚Üí Save as PDF)")
````
